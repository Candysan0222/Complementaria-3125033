```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Ruta del archivo CSV (cambiada a '/content/Data2.csv')
file_path = '/content/Data7.csv'  # Cambiar la ruta según la ubicación de tu archivo

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Convertir los porcentajes a valores decimales (de porcentaje a fracción)
for column in df.columns[1:]:
    df[column] = df[column].str.replace('%', '').astype(float) / 100

# Crear el gráfico
plt.figure(figsize=(14, 8))  # Tamaño del gráfico

# Iterar sobre cada lenguaje para graficar
for index, row in df.iterrows():
    plt.plot(df.columns[1:], row[1:], marker='o', label=row["Language"])

# Ajustar los ticks del eje X para que muestren los meses
plt.xticks(rotation=45, fontsize=12)
plt.xlabel("Meses", fontsize=14)

# Configurar los ticks del eje Y (con decimales redondeados)
plt.yticks(np.linspace(0, 0.2, 11), [f"{int(v * 100)}%" for v in np.linspace(0, 0.2, 11)], fontsize=12)
plt.ylabel("Porcentaje de Uso", fontsize=14)

# Títulos y etiquetas
plt.title("Gráfico Lineal con Marcadores: Lenguajes vs. Meses", fontsize=18)

# Agregar leyenda
plt.legend(title="Lenguajes de Programación", loc='upper left', bbox_to_anchor=(1, 1))

# Cuadrícula para mayor claridad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño y mostrar
plt.tight_layout()
plt.show()
```
