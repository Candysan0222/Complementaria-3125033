```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos desde un archivo CSV
file_path = '/content/Data1.csv'  # Cambia esto por la ruta de tu archivo
data = pd.read_csv(file_path)

# Convertir los datos de porcentaje a valores numéricos
for col in data.columns[1:]:
    data[col] = data[col].str.replace('%', '').astype(float)

# Transformar los datos para graficarlos
data_melted = data.melt(id_vars=["Language"],
                        var_name="Month",
                        value_name="Percentage")

# Configurar el gráfico de dispersión
plt.figure(figsize=(14, 8))
sns.scatterplot(data=data_melted,
                x="Month",
                y="Percentage",
                hue="Language",
                style="Language",
                s=100,
                palette="tab10")

# Personalizar el gráfico
plt.title("Diagrama de correlación de dispersión por lenguaje", fontsize=16)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Porcentaje (%)", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Lenguaje", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
